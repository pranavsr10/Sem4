### Import  Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Load the dataset

```python
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
```

```python
data.keys()
```

```python
print(data['DESCR'])
```

```python
print(data['feature_names'])
```

```python
df = pd.DataFrame(data['data'],columns=data['feature_names'])
df.head(3)
```

### Standardize the data 
>so that the data can have `unit variance` and 
`Standard Deviation = 1`

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
X = sc.fit_transform(df.values)
print(X[0])
```

### Apply the PCA

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
X_pca.shape
```

#### Scatter Plot Insight :
> Below it can be seen that the data has been decomposed from 30 dimentions to 2 PCA components 
which are orthogonal to each other, now having these two components, any appropriate Machine Learning
Classification Algorithm can be used to classify the target classes.

```python
sns.scatterplot(x=X_pca[:,0], y=X_pca[:,1], hue=data['target'])
```

```python
pca_components = pca.components_
df_pca = pd.DataFrame(pca_components,columns=data['feature_names'],index= ['Component:1','Component:2'])
df_pca.head()
```

```python
plt.figure(figsize=(14,5))
sns.heatmap(df_pca.values,cmap='YlGnBu',xticklabels=data['feature_names'])
```

#### Now Applying the Support Vector Classifier to the 2 components

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_pca, data['target'], test_size=0.25, random_state=42)
```

```python
classifier = SVC()
classifier.fit(X_train,y_train)
```

```python
#Making the predictions
preds = classifier.predict(X_test)
```

### Model Evaluation

```python
from sklearn.metrics import confusion_matrix, classification_report,accuracy_score
```

```python
print( 'Accuracy: %.2f' % (accuracy_score(y_test,preds) * 100),'%')
```

```python
cf = pd.DataFrame(confusion_matrix(y_test,preds),columns=['True','False'],index=['True','False'])
cf
```

```python
print(classification_report(y_test,preds))
```
